```python
# Run this in Colab (one cell)
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

data_path = "/content/drive/MyDrive/crypto_project/data"   # your folder

results = []
errors = []

for file in sorted(os.listdir(data_path)):
    if not file.lower().endswith(".csv"):
        continue
    filepath = os.path.join(data_path, file)
    coin_name = file.rsplit(".",1)[0]

    try:
        df = pd.read_csv(filepath)

        # try common column names for price
        price_cols = [c for c in df.columns if c.lower() in ("price","close","close_price","adj_close")]
        if not price_cols:
            # try second method: any numeric column except volume/marketcap
            numeric = df.select_dtypes(include=[np.number]).columns.tolist()
            # prefer column named 'price' else first numeric
            price_col = numeric[0] if numeric else None
        else:
            price_col = price_cols[0]

        if price_col is None:
            raise ValueError("No numeric price column found")

        # date column detection
        date_cols = [c for c in df.columns if "date" in c.lower()]
        if not date_cols:
            # fallback: look for index or a column with yyyy-
            date_col = None
        else:
            date_col = date_cols[0]

        # convert date
        if date_col:
            df['date'] = pd.to_datetime(df[date_col], errors='coerce')
        else:
            # if no date col, try index
            try:
                df.index = pd.to_datetime(df.index)
                df = df.reset_index().rename(columns={'index':'date'})
            except:
                df['date'] = pd.NaT

        df = df.dropna(subset=['date'])
        df = df.sort_values("date").reset_index(drop=True)

        # ensure price is float
        df['price'] = pd.to_numeric(df[price_col], errors='coerce')
        df = df.dropna(subset=['price'])

        # daily returns
        df['daily_return'] = df['price'].pct_change()

        # stats
        daily_vol = df['daily_return'].std()
        annual_vol = daily_vol * np.sqrt(365) if pd.notnull(daily_vol) else np.nan

        # max drawdown
        df['peak'] = df['price'].cummax()
        df['drawdown'] = (df['price'] - df['peak']) / df['peak']
        max_dd = df['drawdown'].min()

        # sharpe (naive, no risk-free rate)
        sharpe = df['daily_return'].mean() / df['daily_return'].std() if df['daily_return'].std() != 0 else np.nan

        results.append([coin_name, daily_vol, annual_vol, max_dd, sharpe])

    except Exception as e:
        errors.append((coin_name, str(e)))
        print("Skipped:", coin_name, "->", e)

# create DataFrame
risk_table = pd.DataFrame(results, columns=[
    "Coin", "Daily Volatility", "Annual Volatility", "Max Drawdown", "Sharpe Ratio"
])

# save to Drive
out_path = "/content/drive/MyDrive/crypto_project/risk_table.csv"
risk_table.to_csv(out_path, index=False)

# show summary
print("Processed:", len(risk_table), "coins")
if errors:
    print("Skipped files:", len(errors))
    for c,err in errors[:10]:
        print(" -", c, ":", err)

# sort and display
risk_table_sorted = risk_table.sort_values("Annual Volatility", ascending=False).reset_index(drop=True)
display(risk_table_sorted.head(20))

# Quick plots (top 8 by annual vol)
topn = risk_table_sorted.head(8).copy()
plt.figure(figsize=(10,5))
plt.bar(topn['Coin'], topn['Annual Volatility'])
plt.title("Top 8 Coins by Annual Volatility")
plt.ylabel("Annual Volatility")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()

plt.figure(figsize=(10,5))
# use absolute drawdown for plotting magnitude
plt.bar(topn['Coin'], topn['Max Drawdown'].abs())
plt.title("Top 8 Coins by Max Drawdown (magnitude)")
plt.ylabel("Max Drawdown (absolute)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
